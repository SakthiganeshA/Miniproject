***5. Top 10 States for Men in Year 2021***

```python
def top_10_states_by_year(df, year, sex):
  df['Date'] = pd.to_datetime(df['Date'], errors='coerce')
  filtered_df = df[
      (df['Date'].dt.year == year) & (df['Sex'] == sex)
  ]
  top_states = filtered_df['State'].value_counts().nlargest(10)
  return top_states
Top_10_states_men_2021 = top_10_states_by_year(df, 2021, 'M')
print(Top_10_states_men_2021)
```

***6. Top 10 States for Women in Year 2021***

```python
def top_10_states_by_year(df, year, sex):
  df['Date'] = pd.to_datetime(df['Date'], errors='coerce')
  filtered_df = df[
      (df['Date'].dt.year == year) & (df['Sex'] == sex)
  ]
  top_states = filtered_df['State'].value_counts().nlargest(10)
  return top_states
Top_10_states_women_2021 = top_10_states_by_year(df, 2021, 'F')
print(Top_10_states_women_2021)
```

***7. Stack bar chart for men vs women X axis -> 18-30 | 31-40 | 41-50 | 51-60***

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
bins = [18, 30, 40, 50, 60]
labels = ['18-30', '31-40', '41-50', '51-60']
df['AgeGroup'] = pd.cut(df['Age'], bins=bins, labels=labels, right=False)
age_sex_counts = pd.crosstab(df['AgeGroup'], df['Sex'])
sns.barplot(x=age_sex_counts.index, y=age_sex_counts['M'], color='blue', label='Men')
sns.barplot(x=age_sex_counts.index, y=age_sex_counts['F'], color='red', bottom=age_sex_counts['M'], label='Women')
plt.title('Stacked Bar Chart for Men vs Women by Age Group')
plt.xlabel('Age Group')
plt.ylabel('Count')
plt.legend()
plt.show()
```

***8. Pie chart for men w.r.t age group (18-30 | 31-40 | 41-50 | 51-60)***

```python
men_age_group = df[df['Sex'] == 'M']['AgeGroup'].value_counts()
plt.figure(figsize=(16,8))
plt.title('Men with refers to Age Group')
plt.piechart = men_age_group.plot.pie(autopct='%1.1f%%')
plt.show()
```

***9. Pie chart for women w.r.t age group (18-30 | 31-40 | 41-50 | 51-60)***

```python
women_age_group = df[df['Sex'] == 'F']['AgeGroup'].value_counts()
plt.figure(figsize=(16,8))
plt.title('Women with refers to Age Group')
plt.piechart = women_age_group.plot.pie(autopct='%1.1f%%')
plt.show()
```
